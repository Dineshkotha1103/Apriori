```python
import numpy as np
import pandas as pd                             #importing libraries
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Market_Basket_Optimisation.csv", header = None)
items = []                                              #data preprocessing
for i in range(0,7501):
  items.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori                              #training the Apriori model on the dataset
rules = apriori(transactions = items, min_support = 0.003, min_confidence = 0.2,min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)                                   #Displaying the first results coming directly from the output of the apriori function
results
```

```python
def inspect(results):                                 #Putting the results well organised into a Pandas DataFrame                       
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
